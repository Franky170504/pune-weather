```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as mlt
```

```python
df_temp = pd.read_csv("data\pune_weather_temperature.csv")
df_temp.head()
```

```python
df_preci = pd.read_csv("data\pune_weather_precipitation.csv")
df_preci.head()
```

```python
# Select only the 'precipitation' column from precipitation CSV
df_preci = df_preci[['tp']]

# Merge both datasets based on index
df_merged = df_temp.join(df_preci, how="left")  # Use "inner" if you want only matching rows

# Save the merged data to a new CSV file
df_merged.to_csv("pune_weather.csv")

print("Merged")
```
